```python
import pandas as pd

# read in the dataset
df = pd.read_csv('/Users/taniadawood/Desktop/DSCI550-PixstoryDatasetAnalysisVisualizations/Datasets/Master_Pixstory_Dataset_EXTRACT_ANALYSIS.csv')

new_df = df.assign(interests=df["Interest"].str.split(",")).explode("Interest")

# group the data by interest and calculate the mean age for each group
interest_age = new_df.groupby("Interest")["Age"].mean()
```

```python
# create an empty dictionary to store the interests and ages
interests_dict = {}

# loop through each row of the DataFrame
for index, row in df.iterrows():
    interests = row["Interest"].split(", ")  # split interests at the ","
    age = row["Age"]
    for interest in interests:
        if interest in interests_dict:
            interests_dict[interest].append(age)
        else:
            interests_dict[interest] = [age]
```

```python
print(interests_dict.keys())
```

```python
import statistics

# create an empty dictionary to store the interest and average age
interests_avg_age = {}

# loop through each key-value pair in the interests_dict
for interest, ages in interests_dict.items():
    avg_age = round(statistics.mean(ages),1) # calculate the average age for this interest
    interests_avg_age[interest] = avg_age  # add the interest and average age to the new dictionary
    
    
    
```

```python
import csv

# create an empty dictionary to store the interest, average age, and count of posts
interests_stats = {}

# loop through each key-value pair in the interests_dict
for interest, ages in interests_dict.items():
    avg_age = round(statistics.mean(ages), 1)  # calculate the average age for this interest and round to 1 decimal place
    count = len(ages)  # count the number of posts for this interest
    interests_stats[interest] = (avg_age, count)  # add the interest, average age, and count to the new dictionary

# write the dictionary to a CSV file
with open('interests_stats.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Interest', 'Average Age', 'Count of Posts'])
    for interest, stats in interests_stats.items():
        avg_age, count = stats
        writer.writerow([interest, avg_age, count])
```
