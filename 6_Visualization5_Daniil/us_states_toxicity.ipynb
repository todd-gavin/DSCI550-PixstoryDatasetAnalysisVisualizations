```python
import pandas as pd
```

```python
df1 = pd.read_csv("../Datasets/Master_Pixstory_Dataset_EXTRACT_ANALYSIS.csv")


df2 = df1.drop(columns=[col for col in df1.columns if col not in ['Date (No Timestamp)', 'GeoTopic Name','toxicity']])

# # Print all column names
print(df1.columns)

df2
```

```python

# Create a list of the 50 US states
us_states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]

# Filter df2 to only show rows that have one of the US 50 states in the "Geotopic Name" column
df2_states = df2[df2['GeoTopic Name'].isin(us_states)]

df2_states
```

```python

# Group df2 by "Geotopic Name" and calculate the mean of the "toxicity" column for each state
df3 = df2_states.groupby("GeoTopic Name")["toxicity"].mean().reset_index()

# Rename the "Geotopic Name" column to "state" for clarity
df3 = df3.rename(columns={"GeoTopic Name": "state"})

df3 = df3.rename(columns={"toxicity": "avg_toxicity"})


#normalize toxicity scores
df3['avg_toxicity'] = df3['avg_toxicity']*1000

# Add a new column called "toxicity_rank" that ranks the "avg_toxicity" column for each state
df3["toxicity_rank"] = df3["avg_toxicity"].rank()

df3
```

```python


df_observable = df3.rename(columns={"avg_toxicity": "rate"})
df_observable = df3.rename(columns={"toxicity_rank": "rank"})
df_observable = df3.rename(columns={"state": "name"})


df_observable.to_csv('us_states_toxicity.csv', index=False)
```
