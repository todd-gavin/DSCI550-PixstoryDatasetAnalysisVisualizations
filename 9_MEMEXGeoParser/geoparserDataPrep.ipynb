```python
# Select specific columns
selected_cols = ['Translated Narrative']
df_geoparser = df[selected_cols]
```

```python
df_geoparser
```

```python
# Randomly sample 1500 rows from the DataFrame and store it in a new DataFrame
df_geoparser_sampled = df_geoparser.sample(n=1500, random_state=42)

# Reset the index and drop the old index column
df_geoparser_sampled.reset_index(drop=True, inplace=True)
```

```python
df_geoparser_sampled
```

```python
# Convert DataFrame to a list of dictionaries
data = [{"text": row["Translated Narrative"], "id": str(index)} for index, row in df_geoparser_sampled.iterrows()]

# Export the list to a JSON file
with open('df_geoparser_sampled_narrative.json', 'w') as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)
```
