```python
# Select specific columns
selected_cols = ['Dataset Index', 'Interest', 'GeoTopic Name', 'GeoTopic Latitude', 'GeoTopic Longitude']
df_geoparser = df[selected_cols]
```

```python
df_geoparser
```

```python
# Drop rows with NaN values in 'GeoTopic Name', 'GeoTopic Latitude', and 'GeoTopic Longitude' columns
df_geoparser = df_geoparser.dropna(subset=['GeoTopic Name', 'GeoTopic Latitude', 'GeoTopic Longitude'])

# Reset index
df_geoparser = df_geoparser.reset_index(drop=True)
```

```python
df_geoparser
```

```python
# First, we need to split the interests in each row into individual elements
df_geoparser['Interest'] = df_geoparser['Interest'].apply(lambda x: x.split(', '))

# Now, we can use the 'explode' function to create new rows for each interest
df_geoparser = df_geoparser.explode('Interest')

# Reset the index and drop the old index column
df_geoparser.reset_index(drop=True, inplace=True)
```

```python
df_geoparser
```

```python
# Randomly sample 1500 rows from the DataFrame and store it in a new DataFrame
df_geoparser_sampled = df_geoparser.sample(n=1500, random_state=42)

# Reset the index and drop the old index column
df_geoparser_sampled.reset_index(drop=True, inplace=True)
```

```python
df_geoparser_sampled
```

```python
# Drop the "Dataset Index" column from the DataFrame
df_geoparser_sampled.drop('Dataset Index', axis=1, inplace=True)
```

```python
df_geoparser_sampled
```

```python
# Export the DataFrame to a JSON file
df_geoparser_sampled.to_json('df_geoparser_sampled.json', orient='records')
```
